```python
import bs4
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
```

```python
bs4.SoupStrainer(
    'div',
    attrs = {'class' : ['newsct_article _article_body', 'media_end_head_title']}
)
```

```python
loader = WebBaseLoader(
    web_path = ('https://n.news.naver.com/article/437/0000420199?sid=101'),
    bs_kwargs = dict(
        parse_only = bs4.SoupStrainer(
            'div',
        attrs = {'class' : ['newsct_article _article_body', 'media_end_head_title']}
        )       
    )
)
docs = loader.load()
```

```python
print(len(docs))
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
```

```python
splits = text_splitter.split_documents(docs)
len(splits)
```

```python
vectorstore = FAISS.from_documents(documents=splits, embedding=OpenAIEmbeddings())
# ë²¡í„° db ìƒì„±
```

```python
retriever = vectorstore.as_retriever()
# ê²€ìƒ‰ê¸° ìƒì„±
```

```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template(
    """
    ë‹¹ì‹ ì€ ì§ˆë¬¸ì— ë”°ë¼ ë‹µë³€ì„ ìˆ˜í–‰í•˜ëŠ” ì¹œì ˆí•œ AI ë¹„ì„œ ì…ë‹ˆë‹¤. ë‹¹ì‹ ì€ ì£¼ì–´ì§„ contectì—ì„œ ì£¼ì–´ì§„ questionì— ë‹µí•˜ëŠ” ê²ƒì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.
    ê²€ìƒ‰ëœ ê²°ê³¼ì¸ ë‹¤ìŒ contextë¥¼ ì‚¬ìš©í•˜ì—¬ ì§ˆë¬¸ì¸ questionì— ë‹µí•˜ì„¸ìš”.
    ë§Œì•½, contextì—ì„œ ë‹µì„ ì°¾ì„ ìˆ˜ ì—†ê±°ë‚˜, ë‹µì„ ëª¨ë¥¸ë‹¤ë©´ 'ì£¼ì–´ì§„ ì •ë³´ì—ì„œ ì§ˆë¬¸ì— ëŒ€í•œ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.' ë¼ê³  í•œê¸€ë¡œ ë‹µë³€í•´ ì£¼ì„¸ìš”.
    ì´ë¦„ì´ë‚˜ ê¸°ìˆ ì ì¸ ìš©ì–´ëŠ” ë²ˆì—­í•˜ì§€ ì•Šê³  ê·¸ëŒ€ë¡œ ì¶œë ¥í•´ ì£¼ì„¸ìš”.

    # Question : {question}
    # Context : {context}
    # Answer : 
    """
)
```

```python
llm = ChatOpenAI(model = 'gpt-4o', temperature = 0)
```

```python
rag_chain = (
    {'context' : retriever, 'question' : RunnablePassthrough()}
    | prompt
    | llm
    |StrOutputParser() # ì¶œë ¥ì„ ê¹”ê¼¼í•˜ê²Œ ì •ë¦¬!!!!!!!!!!!
)
```
