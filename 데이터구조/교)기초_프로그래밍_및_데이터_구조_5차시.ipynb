````python
# Q2. íŒ©í† ë¦¬ì–¼ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ë³´ì„¸ìš”.
# factorial(5)ì˜ ì¶œë ¥ ì˜ˆì‹œ : 5! = 1 * 2 * 3 * 4 * 5 = 120

def factorial(num):
    a = []
    while num > 0:       
            a.append(num)
            num -= 1
    print(a)
    result = 1
    for i in a:
          result *= i
    return print(f'{num}! = {a[:-1]} = {result}')

factorial(5)
````

````python
def factorial(n):
    result = 1
    cal = ''
    for i in range(1, n+1):
        result *= i
        if i < n:
            cal += str(i) + ' * '
        else:
            cal += str(i)
    print(f'{n}! = {cal} = {result}')

factorial(5)
````
