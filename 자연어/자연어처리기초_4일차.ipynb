#### 1. RNNÏùÑ Ïù¥Ïö©ÌïòÏó¨ ÌÖçÏä§Ìä∏ ÏÉùÏÑ±ÌïòÍ∏∞

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
text = """Í≤ΩÎßàÏû•Ïóê ÏûàÎäî ÎßêÏù¥ Îõ∞Í≥† ÏûàÎã§\n
Í∑∏Ïùò ÎßêÏù¥ Î≤ïÏù¥Îã§\n
Í∞ÄÎäî ÎßêÏù¥ Í≥†ÏôÄÏïº Ïò§Îäî ÎßêÏù¥ Í≥±Îã§\n"""
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
sequences = []
for line in text.split('\n'):
    encoded = tokenizer.texts_to_sequences([line])[0]
    # print(encoded)
    for i in range(1, len(encoded)):
        # i = 1Ïùº Îïå, sequence = [2,3]
        # i = 2Ïùº Îïå, sequence = [2,3,1]
        sequence = encoded[:i+1]
        sequences.append(sequence)

print(f'ÌïôÏäµÏóê ÏÇ¨Ïö©Ìï† ÏÉòÌîåÏùò Í∞úÏàò : {len(sequences)}')
```

```python
sequences # ÏàúÏ∞®Ï†ÅÏúºÎ°ú ÎßåÎì§Í∏∞
```

```python
# Î¨∏Ïû•Ïùò Í∏∏Ïù¥Î•º ÎèôÏùºÌïòÍ≤å ÎßûÏ∂îÍ∏∞

lengths = []
for i in sequences:
    lengths.append(len(i))
```

```python
max_len = max(lengths)
print(f'ÏÉòÌîå Ï§ë ÏµúÎåÄ Í∏∏Ïù¥ : {max_len}')
```

```python
sequences = pad_sequences(sequences, maxlen=max_len, padding = 'pre') # pad_sequences : Í∏∏Ïù¥Î•º ÎßûÏ∂îÍ∏∞ ÏúÑÌï¥ Ìå®Îî©ÏùÑ ÎÑ£Ïñ¥Ï£ºÎäî Ìï®Ïàò
```

```python
sequences
```

```python
sequences = np.array(sequences) # Ïò§Î•ò Î∞úÏÉù Ìï† Ïàò ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê 
```

```python
X = sequences[:, :-1]
y = sequences[:, -1]
```

```python
X
```

```python
y
```

```python
y = to_categorical(y, num_classes=vocab_size) # one-hot incoding
y
```

#### 2) Î™®Îç∏ ÏÑ§Í≥ÑÌïòÍ∏∞

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN # Embedding : Îã®Ïñ¥Ï†Å ÏùòÎØ∏Î•º Ìè¨Ìï®ÌïòÍ≥† ÏûàÎã§

embedding_dim = 10
hidden_units = 32
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(vocab_size, activation='softmax'))
# word_index = 11, vector(padding) = 12, 
# vocab_size = 12Ïù∏Îç∞ Îã®Ïñ¥ 11Í∞ú + 1 ÌñàÎçò word_indexÍ∞í
# +1Ïù∏ ÏûêÎ¶¨Ï∞®ÏßÄÎ•º ÏúÑÌï¥ ÏïûÏóê 0ÏùÑ Ï±ÑÏõåÏ§Ä Í≤ÉÏúºÎ°ú
# Ïã§Ï†ú ÏòàÏ∏°ÏùÑ ÏàòÌñâÌï† ÎïåÏóêÎäî 0ÏùÄ Îã®Ïñ¥Í∞Ä ÏïÑÎãàÎØÄÎ°ú Î¨¥ÏãúÎêòÏñ¥ Ï∂úÎ†• Í∞íÏúºÎ°ú 0Ïù¥ ÎÇòÏò§ÏßÄÎäî ÏïäÎäîÎã§

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, y, epochs=200, verbose=2)

```

```python
# Î¨∏Ïû•ÏùÑ ÏÉùÏÑ±ÌïòÎäî Ìï®Ïàò ÎßåÎì§Í∏∞
def sentence_generation(model, tokenizer, current_word, n):
    init_word = current_word # ÏÇ¨Ïö©ÏûêÍ∞Ä ÏûÖÎ†•Ìïú Îã®Ïñ¥
    sentence = '' # ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú ÏÉùÏÑ±Ìïú Î¨∏Ïû•ÏùÑ Î∞òÌôò

    for i in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=5, padding='pre') # maxlen= : inputÍ≥º outputÏùò Í∏∏Ïù¥Í∞Ä Í∞ôÏïÑÏïº ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê 5

        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis=1)

        for word, index in tokenizer.word_index.items():
            if index == result:
                break
            # (ÎßêÏù¥ ,1)
        current_word = current_word + ' ' + word
        # current_wordÎäî Îã§Ïùå Î∞òÎ≥µÎ¨∏ÏóêÏÑú ÏÇ¨Ïö©ÎêòÍ∏∞ ÏúÑÌï¥ Ï†ÄÏû•Ìïú Î≥ÄÏàò
        sentence = sentence + ' ' + word
        # sentenceÎäî ÌòÑÏû¨ Ï∂úÎ†•ÏùÑ ÏúÑÌïú Î¨∏Ïû• ÏÉùÏÑ±
    sentence = init_word + sentence
    return sentence
```

#### Ï†ÑÏ≤òÎ¶¨

```python
# Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò
def repreprocessing(raw_sentence):
    preprocessed_sentence = raw_sentence.encode('utf8').decode('ascii', 'ignore')
    # ÏïÑÏä§ÌÇ§ : Îã®ÏàúÌïòÍ≥† ÏòÅÏñ¥ ÌÖçÏä§Ìä∏Ïóê Ìö®Í≥ºÏ†Å
    # utf8 : Îã§Íµ≠Ïñ¥ÏôÄ Í∏∞Ìò∏ Ï≤òÎ¶¨ Í∞ÄÎä•
    # ignore : ÏïÑÏä§ÌÇ§ Î¨∏ÏûêÍ∞Ä ÏïÑÎãåÍ≤ÉÎì§ÏùÄ Î¨¥ÏãúÌïúÎã§
    # Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ ÏòÅÏñ¥Î°ú ÎêòÏñ¥ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê Ïö∞ÏÑ†Ï†ÅÏúºÎ°ú Îã§Íµ≠Ïñ¥ÏôÄ Í∏∞Ìò∏Î•º Ï≤òÎ¶¨Ìïú Îí§
    # Îã®ÏàúÌïòÍ≥† ÏòÅÏñ¥ ÌÖçÏä§Ìä∏Ïóê Ìö®Í≥ºÏ†ÅÏù∏ ÏïÑÏä§ÌÇ§ Î¨∏ÏûêÎ°ú Î≥ÄÌôò
    result = ''
    for word in preprocessed_sentence:
        if word not in punctuation: # Íµ¨ÎëêÏ†ê Ï†úÍ±∞
            result += word
    return result.lower() # ÏÜåÎ¨∏Ïûê Ìôî
```

```python
preprocessed_headline = []
# Ï†ÑÏ≤òÎ¶¨ Îêú Ìó§ÎìúÎùºÏù∏ Îã¥ÏïÑÏ£ºÍ∏∞
for word in clean_headline:
    preprocessed_sentence = repreprocessing(word)
    preprocessed_headline.append(preprocessed_sentence)

preprocessed_headline[:5]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_headline)
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(f'Îã®Ïñ¥ ÏßëÌï©Ïùò ÌÅ¨Í∏∞ : {vocab_size}')
```

```python
sequences = []
for sentence in preprocessed_headline:
    encoded = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        # Ïù∏ÏΩîÎî©Îêú Î¨∏Ïû•Ïù¥ [3, 5, 6, 7] ÎùºÎ©¥ 
        # Ï≤´Î≤àÏß∏ Î∞òÎ≥µÏóêÏÑúÎäî [3, 5] 
        # ÎëêÎ≤àÏß∏ Î∞òÎ≥µÏóêÏÑúÎäî [3, 5, 6]
        sequences.append(sequence)
```

#### 2) Î™®Îç∏ ÏÑ§Í≥ÑÌïòÍ∏∞

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM

embedding_dim = 10 # Í∞ÅÍ∞ÅÏùò Îã®Ïñ¥Îì§Ïùò ÏùòÎØ∏Î•º Îã¥Í≥† ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê ÏùòÎØ∏Î•º Ï∞®ÏõêÏóê Îã¥ÏïÑ Îë† (10ÏúºÎ°ú Îëî Ïù¥Ïú†Îäî ÏãúÍ∞ÑÎì±...Ïùò Î¨∏Ï†úÎ°ú Ïù∏Ìï¥)
# embedding_dimÏù¥ ÌÅ¥ ÏàòÎ°ù Îçî ÏÑ∏Î∞ÄÌïú ÌëúÌòÑ Í∞ÄÎä•
hidden_units = 128

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(LSTM(hidden_units))
model.add(Dense(vocab_size, activation='softmax')) # softmax: ÌôïÎ•†ÏùÑ Íµ¨ÌïòÍ≥† Îã§ ÎçîÌïòÎ©¥ 1

# Î™®Îç∏Ïù¥ ÌïôÏäµÌï† Ï§ÄÎπÑ
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
# Binary Cross-Entropy (Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº)
# Categorical Cross-Entropy (Îã§Ï§ë ÌÅ¥ÎûòÏä§ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº)
model.fit(X, y, epochs = 200, verbose=2)
```

```python
def sentence_generation(model, tokenizer, current_word, n):
    init_word = current_word
    sentence = ''

    for i in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre') # ÌïúÎã®Ïñ¥ ÏçºÍ∏∞ ÎïåÎ¨∏Ïóê -1

        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis = 1)
        # Îã®Ïñ¥ÏßëÌï©ÏÇ¨Ï†ÑÏóêÏÑúÏùò Ïù∏Îç±Ïä§Í∞íÏúºÎ°ú resultÍ∞Ä Ï†ÄÏû•Îê®
        # -> ÏòàÏ∏°Ìïú Îã®Ïñ¥Î°ú Í∞ÄÏû• ÌôïÎ•†Ïù¥ ÎÜíÏùÄ Îã®Ïñ¥Ïùò Ïù∏Îç±Ïä§ Í∞í

        #tokenizer.word_index -> {Îã®Ïñ¥ : Ïù∏Îç±Ïä§, Îã®Ïñ¥ : Ïù∏Îç±Ïä§}
        for word, index in tokenizer.word_index.items(): # .items() : Î™®Îì† ÌÇ§ÏôÄ Í∞íÏù¥ ÌäúÌîåÌòïÌÉúÎ°ú ÎÇòÏò¥
            if index == result:
                break

        current_word = current_word + ' ' + word
        sentence = sentence + ' ' + word
    sentence = init_word + sentence
    return sentence
```
