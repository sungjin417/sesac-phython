## 3. í…ìŠ¤íŠ¸ ë²¡í„°í™”

```python
def bag_of_words(text):
    okt = Okt()
    text = text.replace('.', '') # ë¬¸ì¥ì˜ . ì„ ì²˜ë¦¬
    tokens = okt.morphs(text)
    word_dict = {}
    bow = []

    for token in tokens:
        if token not in word_dict:
            word_dict[token] = len(word_dict) # ì¸ë±ìŠ¤ ë¶€ì—¬
            bow.append(1)
        else:
            index = word_dict[token]
            bow[index] += 1
    return word_dict, bow
```

```python
doc = 'ë°˜ë³µë˜ëŠ” ë‹¨ì–´ëŠ” ë‹¨ì–´ ê°œìˆ˜ ë§Œí¼ ì¶”ê°€ ë©ë‹ˆë‹¤..'
```

```python
word_dict, bow = bag_of_words(doc)
print(bow)
print(word_dict)
```

```python
import pandas as pd
from math import log
# DTM : ë¬¸ì„œ ë‹¨ì–´ í–‰ë ¬ (Document Term Matrix)

docs = [
  'ì›ìˆ­ì´ ì—‰ë©ì´ëŠ” ë¹¨ê°œ',
  'ë¹¨ê°„ê±´ ì‚¬ê³¼',
  'ì‚¬ê³¼ëŠ” ë§›ìˆì–´',
  'ë§›ìˆìœ¼ë©´ ë°”ë‚˜ë‚˜'
]
```

```python
vocab = []
for doc in docs:
    words = doc.split() # ë‹¨ì–´ë‹¨ìœ„ë¡œ ë¶„ë¦¬(ê³µë°±)
    for w in words:
        if w not in vocab:
            vocab.append(w)
vocab.sort()
vocab
```

```python
N = len(docs)
print(N)
def tf(t, d):
    return d.count(t)
# TF : íŠ¹ì • ë¬¸ì„œ dì—ì„œ íŠ¹ì • ë‹¨ì–´ tì˜ ë“±ì¥ íšŸìˆ˜
```

```python
# ë¬¸ì„œ ì§‘í•©ì—ì„œ ì–¼ë§ˆë‚˜ í”.í•˜.ì§€ ì•Šì€ì§€
# df : ë¬¸ì„œì˜ ë¹ˆë„
# docs : ë¬¸ì¥ì˜ ì§‘í•©
# doc : í•œ ë¬¸ì¥ì”© ë°˜í™˜
# t : ë‹¨ì–´
# N : ë¬¸ì¥ì˜ ì§‘í•©ì•ˆì˜ ë¬¸ì¥ì˜ ìˆ˜

def idf(t):
    df = 0 # ë¬¸ì„œì˜ë¹ˆë„ë¥¼ ì €ì¥(t ê°€ ë“±ì¥í•˜ëŠ” ë¬¸ì„œì˜ ê°œìˆ˜)
    for doc in docs:
        df += t in doc # True(1), False(0)ë¡œ ë°˜í™˜ ë¨
    return log((N/df+1)) # df + 1ë¡œ ë‚˜ëˆ„ëŠ” ê²ƒì€ 0ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” ì˜¤ë¥˜ë¥¼ ë°©ì§€í•˜ê¸° ìœ„í•´
```

```python
def tfidf(t, d): 
    return tf(t,d) *idf(t)
```

```python
result = []
for i in range(N):
    result.append([])
    d = docs[i]

    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t,d))

tf_ = pd.DataFrame(result, columns= vocab)
tf_
```

```python
restaurants['char_count'] = restaurants['Review'].apply(lambda x: len(x.replace(" ", "")))
restaurants
```

```python
# restaurants['Review']
res_char = restaurants.copy()
res_word = restaurants.copy()
```

```python
char_list = []
for char in res_char['Review']:
    char_list.append(len(char.replace(" ", "")))
res_char['char'] = char_list    
res_char
```

```python
word_list = []
for word in res_word['Review']:
    words = word.split()
    word_list.append(len(words))
res_word['word'] = word_list
res_word.iloc[0,1]       
res_word
```

```python
# í‰ì ì— ë”°ë¥¸ ê¸¸ì´ì´ì´ì´

word = res_word.groupby('Rating')['word'].mean()
char = res_char.groupby('Rating')['char'].mean()
print(f'ë‹¨ì–´ì˜ í‰ê·  : {word}, \në¬¸ì¥ê¸¸ì´ì˜ í‰ê·  : {char}')

```
