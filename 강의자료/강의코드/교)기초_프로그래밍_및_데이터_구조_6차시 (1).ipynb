```python
# Q1. ì‚¬ìš©ìì—ê²Œ ì›ì˜ ë°˜ì§€ë¦„ì„ ì…ë ¥ë°›ì•„ ì›ì˜ ë‘˜ë ˆì™€ ì›ì˜ ë„“ì´ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ë³´ì„¸ìš”.
# ì›ì˜ ë‘˜ë ˆ : 2 * ì›ì£¼ìœ¨ * ì›ì˜ ë°˜ì§€ë¦„
# ì›ì˜ ë„“ì´ : ì›ì£¼ìœ¨ * ì›ì˜ ë°˜ì§€ë¦„ì˜ ê±°ë“­ì œê³±
# ì›ì˜ ë‘˜ë ˆì™€ ë„“ì´ëŠ” ì†Œìˆ˜ì  2ë²ˆì§¸ ìë¦¬ìˆ˜ê¹Œì§€ë§Œ ì¶œë ¥í•©ë‹ˆë‹¤ 

# import math

def circle(num):
    answer = 2 * math.pi * num # ì›ì˜ ë‘˜ë ˆ
    answer2 = math.pi * num ** 2
    return answer, answer2
```

```python
radius = float(input('ì›ì˜ ë°˜ì§€ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” : '))
ë‘˜ë ˆ, ë„“ì´ = circle(radius)
print(f'ì›ì˜ ë‘˜ë ˆ : {ë‘˜ë ˆ:.2f}, ì›ì˜ ë„“ì´ : {ë„“ì´:.2f}')
```

```python
# Q2. ì‚¬ìš©ìì—ê²Œ íƒœì–´ë‚œ ë…„ë„ì™€ ë‹¬ì„ ì…ë ¥ë°›ì•„ ì˜¤ëŠ˜ ë‚ ì§œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚˜ì´ê°€ ëª‡ ì„¸ì¸ì§€ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ë³´ì„¸ìš”
# 1. íƒœì–´ë‚œ ë…„ë„ì™€ ë‹¬ì„ ì…ë ¥ë°›ì•„ ì˜¤ëŠ˜ ë‚ ì§œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì˜¤ëŠ˜ì˜ year-íƒœì–´ë‚œ ì—°ë„
# monthë¥¼ ë¹„êµí•˜ì—¬ ìƒì¼ì´ ì§€ë‚˜ì§€ ì•Šì•˜ì„ ê²½ìš° ë‚˜ì´ì—ì„œ 1ì„ ë¹¼ê³ ,
# ìƒì¼ ë‹¬ê³¼ í˜„ì¬ ë‚ ì§œì˜ ë‹¬ì´ ê°™ê±°ë‚˜ ì§€ë‚¬ë‹¤ë©´ ë¹¼ì§€ ì•ŠìŠµë‹ˆë‹¤.

def age(year, month):
    today = datetime.datetime.now()
    age = today.year - year

    if today.month < month:
        age -= 1
    elif today.month >= month:
        pass
    return age
```

```python
birth_year = int(input('ì¶œìƒ ì—°ë„ë¥¼ ë„¤ ìë¦¬ ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”'))
birth_month = int(input('ì¶œìƒ ì›”ì„ ì…ë ¥í•´ì£¼ì„¸ìš”'))
current_age = age(birth_year, birth_month)
print(current_age)
```
