```python
import numpy as np
import random
import tensorflow as tf
from transformers import AutoTokenizer
from transformers import TFGPT2LMHeadModel
```

```python
model = TFGPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained('skt/kogpt2-base-v2')
```

```python
sent = 'ì¸ìƒì— ëŒ€í•˜ì—¬'
```

```python
input_ids = tokenizer.encode(sent)
```

```python
input_ids
```

```python
input_ids = tf.convert_to_tensor([input_ids])
# ì •ìˆ˜ ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜
```

```python
input_ids.shape
# 2ì°¨ì› í…ì„œë¡œ ë³€í™˜ ë˜ì—ˆìŒ
# (1,3) : (ì…ë ¥ë°ì´í„°ì˜ ë°°ì¹˜ í¬ê¸°, ë¬¸ì¥ì˜ í† í° ìˆ˜)
```

```python
output = model.generate(input_ids, max_length = 128, repetition_penalty = 2.0, use_cache = True)
# repetition_penalty = 2.0 : ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©
# repetition_penalty = 1.0 : íŒ¨ë„í‹°ê°€ ì—†ìŒ
# use_cache = True : ì „ì— ì‚¬ìš©í–ˆë˜ ë¬¸ì¥ì„ ìƒì„±í•˜ëŠ”ë° ì‚¬ìš©í–ˆë˜ ìºì‹œ ì‚¬ìš©
```

```python
sent = 'ë‚´ì¼ ì•„ì¹¨ì—'
input_ids = tokenizer.encode(sent)
# ì…ë ¥ ë¬¸ì¥ì„ ì •ìˆ˜ ì¸ë±ìŠ¤ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜ í›„ input_idsì— ì €ì¥

while len(input_ids) < 50:
  # í…ìŠ¤íŠ¸ ì´ ê¸¸ì´(í† í°ê°œìˆ˜)ê°€ 50ê°œ ë¯¸ë§Œì¸ ë™ì•ˆ ìƒˆë¡œìš´ ë‹¨ì–´ ì¶”ê°€
  output = model(np.array([input_ids]))
  # í˜„ì¬ê¹Œì§€ ìƒì„±ëœ í† í°(input_ids)ë¥¼ ëª¨ë¸ì— ì…ë ¥ìœ¼ë¡œ ì œê³µ
  top5 = tf.math.top_k(output.logits[0, -1], k=5)
  # top5 ë‹¨ì–´ë“¤ì„ ì¶”ì¶œ
  token_id = random.choice(top5.indices.numpy())
  # top5 ë‹¨ì–´ë“¤ ì¤‘ ëœë¤ìœ¼ë¡œ ë‹¤ìŒ ë‹¨ì–´ ì„ íƒ
  # ëª¨ë¸ ì¶œë ¥ì˜ ë‹¤ì–‘ì„±ì„ í™•ë³´í•˜ê¸° ìœ„í•¨
  input_ids.append(token_id)
  # input_idsì— token_id ì €ì¥
tokenizer.decode(input_ids)
# ì •ìˆ˜ ì¸ë±ìŠ¤ë¥¼ í•œê¸€ë¡œ decodingí•˜ì—¬ ì¶œë ¥
```
